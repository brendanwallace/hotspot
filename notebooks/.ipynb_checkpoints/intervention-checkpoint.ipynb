```python
def percentile(trial_result, percent=0.8, reverse=False):
    EXTINCTION_THRESHOLD = 10
    secondaries_flat = []
    for sec_ in trial_result['SimulationResults']['SecondaryInfectionCounts']:
        if sum(sec_) > EXTINCTION_THRESHOLD:
            secondaries_flat.extend(sec_)
    secondaries_flat = sorted(secondaries_flat, reverse=True)

    secondaries, primaries = [], []
    secondaries_sum, primaries_sum = 0, 0
    for s in secondaries_flat:
        secondaries_sum += s
        primaries_sum += 1
        secondaries.append(secondaries_sum)
        primaries.append(primaries_sum)

    secondaries = [s / float(secondaries_sum) for s in secondaries]
    primaries = [p / float(primaries_sum) for p in primaries]

    # find 80% mark
    s = 0
    if reverse:
        while primaries[s] < percent:
            s += 1
        return(secondaries[s])
    else:
        while secondaries[s] < percent:
            s += 1
        return(primaries[s])


def plot_trialresults(TrialResults):
    for TrialResult in TrialResults:
        secondaries_flat = []
        for sec_ in TrialResult['SimulationResults']['SecondaryInfectionCounts']:
            secondaries_flat.extend(sec_)
        secondaries_flat = sorted(secondaries_flat, reverse=True)

        secondaries, primaries = [], []
        secondaries_sum, primaries_sum = 0, 0
        for s in secondaries_flat:
            secondaries_sum += s
            primaries_sum += 1
            secondaries.append(secondaries_sum)
            primaries.append(primaries_sum)

        secondaries = [s / float(secondaries_sum) for s in secondaries]
        primaries = [p / float(primaries_sum) for p in primaries]
        
        # find 80% mark
        s = 0
        while secondaries[s] < 0.8:
            s += 1
        print(primaries[s])
        
        
        print(TrialResult['Parameters'])
        plt.subplot(1, 2, 1)
        plt.hist(secondaries_flat)

        plt.subplot(1, 2, 2)
        plt.plot(primaries, secondaries)
        plt.axvline(x=0.15, color='black', linestyle=':')
        plt.axvline(x=0.2, color='black', linestyle='-')
        
#         plt.subplot(1, 3, 3)
#         x = np.linspace(alpha.ppf(0.01, a), alpha.ppf(0.99, a), 100)
#         plt.plot(x, alpha.pdf(a, ))
        
        
        plt.show()
```

```python
for R0, RISK_ in RISK_STRUCTURED.items():
    for i, p in enumerate(['1', '25', '5', '75']):
        for j, var in enumerate(['low', 'medium', 'high']):
            series = RISK_[(2-i)*3 + (2-j)]
    
        

        
            alphaCs = []
            values = []
            for tr in series['TrialResults'][:-1]:
                alphaCs.append(tr['Parameters']['AlphaC']*1000)
                #alphaCs.append(tr['Parameters']['AlphaR']*1000)
                values.append(percentile(tr, percent=0.2, reverse=True))
            plt.plot(alphaCs, values, label=series['Description'], color=colors[i], linestyle=linestyles[j])
    plt.legend()
    plt.gca().invert_xaxis()
    plt.show()
```
