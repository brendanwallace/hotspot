```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
R0 = 18
epsilon = 0.0001
time_limit = 10
granularity = 10000
T = np.linspace(0, time_limit, granularity)
X0 = [1 - epsilon, epsilon, 0]

print(f'R0={R0}')

def model(X, _):
    S, I, R = X
    dSdt = -(R0 * S * I)
    dIdt = (R0 * S * I) - I
    dRdt =  I
    
    return [dSdt, dIdt, dRdt]

solution = odeint(model, X0, T)
plt.plot(T, solution[:, 0], label="Susceptible")
plt.plot(T, solution[:, 1], label="Infected")
plt.plot(T, solution[:, 2], label='"Recovered"')
plt.xlabel('disease generations')
plt.ylabel('population fraction')
plt.title(f'SIR model dynamics R0={R0}')
plt.legend()
```
