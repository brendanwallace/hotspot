```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
```

```python
a = 1.5
b = 3

x = np.linspace(0, 1, 100)
plt.plot(x, beta.cdf(x, a, b))
plt.plot(x, 0*x, c='black')
plt.plot(x, beta.pdf(x, a, b))
plt.ylim([0, 5])
```
