```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
```

```python
FILELOCATION = "/Users/brendan/Documents/projects/epidemiology_risky_nonrisky/go/data/"
UNIFORM = "uniform,T=10000,N=1000,9:21AM.json"
DOUBLE_AR = "uniform,T=1000,N=1000,ac=0.0001,ar=0.0008,dl=10,7:47AM.json"
TRIPLE_AR = "uniform,T=1000,N=1000,ac=0.0001,ar=0.0012,dl=10,8:02AM.json"
QUAD_AR = "uniform,T=1000,N=1000,ac=0.0001,ar=0.0016,dl=10,8:05AM.json"

def imshow_events_by_p_time(data, mag=10):
    flat = [item for l in data['InfectionEvents'] for item in l]
    events = pd.DataFrame(pd.json_normalize(flat))
    N = data['Parameters']['N']
    space = np.zeros((N//mag, N//mag))
    for _, event in events.iterrows():
        r = int(np.floor(event['InfecteeRiskyness'] * (N//mag)))
        #t = min(event['EventTime.Steps']//2, N//mag-1)
        t = (N - event['EventTime.Succeptible'])//mag
        space[r, t] += 1
    plt.imshow(space)
    plt.colorbar()
    plt.show()

        
def _peak(series):
    series = np.array(series)
    m = max(series)
    return np.where(series == m)[0][0]
    
def plot_events_by_time(data):
    """
    rescale the time by peak infection (since these fall so far apart)
    """
    N = data['Parameters']['N']
    Rs_total = np.zeros(N)
    Cs_total = np.zeros(N)
    latest_peak = -1
    for infections in data['InfectionEvents']:
        Rs = np.zeros(N)
        Cs = np.zeros(N)
        for event in infections:
            if event['WasRiskyEvent']:
                Rs[event['EventTime.Steps']] += 1
            else:
                Cs[event['EventTime.Steps']] += 1
        
        Totals = Rs + Cs
        peak = _peak(Totals)
        if peak <= latest_peak:
            np.roll(Rs, latest_peak - peak)
            np.roll(Cs, latest_peak - peak)
        else:
            np.roll(Rs_total, peak - latest_peak)
            np.roll(Cs_total, peak - latest_peak)
            latest_peak = peak
        Rs_total += Rs
        Cs_total += Cs
    
    T = 0
    while Rs_total[T] > 0 or Cs_total[T] > 0:
        T += 1

    plt.plot(Rs_total[:T], label="risky infections")
    plt.plot(Cs_total[:T], label="community infections")
    plt.plot(Cs_total[:T]+Rs_total[:T], label="sum")
    plt.legend()
    plt.show()


def analyze(filename, Rhist=True, timehist=True, imshow=True, plot_by_time=True):
    with open(FILELOCATION+filename, "r") as file:
        data = json.loads(file.read())
    print(data['Parameters'])
    
    # histogram of final R
    plt.hist(data['FinalRs'], bins=40)
    plt.title("histogram of final R (i.e. epidemic size)")
    plt.show()
    
    # histogram of time of peak

    plt.hist([_peak(series) for series in data['Infecteds']], bins=30)
    plt.title('histogram of peak of epidemic')
    plt.show()

    if imshow:
        imshow_events_by_p_time(data)
        
    if plot_by_time:
        plot_events_by_time(data)
```

```python
analyze("uniform,T=1000,N=1000,ac=0.0002,ar=0.0016,dl=10,10:19AM.json", imshow=False)
```
