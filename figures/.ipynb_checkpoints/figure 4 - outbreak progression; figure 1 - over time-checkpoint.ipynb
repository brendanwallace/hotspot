```python
import numpy as np
import json
from scipy import stats
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

%matplotlib inline


DPI = 300

DATA_LOCATION =  '/Users/brendan/Documents/projects/problem_place/go/data/'
IMAGE_LOCATION = '/Users/brendan/Documents/projects/problem_place/writeup/images/abm_first'


EXTINCTION_CUTOFF = 50
TRIALS = 1000
N = 1000

## Plotting setup
#b0c0f7
#

cm = 1/2.54
DPI = 600
colors = ["black","#be89f0", "#f74a64"]
#colors = ["#466be3","#be89f0", "#f74a64"]

sns.set_style("ticks")
sns.set_palette(colors)
sns.color_palette()
sns.set_context("paper", rc={"font.size":12,"axes.titlesize":12,"axes.labelsize":12})   
```

```python
with open(DATA_LOCATION + "R0Series,g=1.0,T=1.json") as file:
    json_file = json.load(file, parse_float=lambda f: round(float(f), 2))
    
R0Series = pd.json_normalize(
json_file,
record_path=["RunSets", "Runs"],
meta=[
    "RiskMean",
    "RiskVariance",
    "ProblemPlaceFraction",
    ["RunSets", "Parameters", "R0"],
    ["RunSets", "Parameters", "RunType"],
],
    errors="ignore",
)#.drop(columns=["Is", "Rts", "EffectiveAlphas", "IRisks", "SRisks"])

R0Series["Extinction Probability"] = 1.0*(R0Series["FinalR"] <= EXTINCTION_CUTOFF)
R0Series["RiskMeanCat"] = pd.Categorical(R0Series["RiskMean"])

```

```python
#Sim = R0Series[(R0Series["RunSets.Parameters.RunType"] == "simulation") & (R0Series["ProblemPlaceFraction"] != 0)]
DifEq = R0Series[(R0Series["RunSets.Parameters.RunType"] == "difeq")]
#Difference = R0Series[(R0Series["RunSets.Parameters.RunType"] == "difference") & (R0Series["ProblemPlaceFraction"] != 0)]
```

```python
column_names = {
    "RunSets.Parameters.R0": "R0",
    "ProblemPlaceFraction": "Hot Spot Fraction",
    "RiskVariance": "Risk Distribution Variance",
    "RiskMeanCat": "Risk Distribution Mean",
    "MaxIDiff": "Peak Size Difference",
    "FinalRDiff": "Extent Difference",
}

DifEq = DifEq.rename(columns=column_names)
```

```python
DifEq["RiskMean"]
```

```python
Study = DifEq[
    (DifEq["RiskMean"] == 0.25) &
    (DifEq["Risk Distribution Variance"] == "high")]


Study = Study[(Study["R0"] == 1.2) | (Study["R0"] == 2.0) | (Study["R0"] == 3.0)]
Study = Study[(Study["Hot Spot Fraction"] == 0.0) | (Study["Hot Spot Fraction"] == 0.5)]


Study = Study.explode(["Ts", "Is", "Rs", "Rts"])
mask = Study["Hot Spot Fraction"] == 0.5
Study["Model Type"] = pd.Categorical(np.where(mask, "Hot Spot", "Homogeneous"))
```

```python
def plot_row(y, ylabel=None, legend=False):
    g = sns.FacetGrid(
        data=Study, col="R0", hue="Model Type", sharex=False, sharey=True,
        height=11*cm*2/3,
        aspect=1.0,
    ) # model_type hue
    g.map_dataframe(sns.lineplot, x="Ts", y=y, hue="Model Type", estimator=None, n_boot=0)
    g.set(ylabel=ylabel)
    g.set(xlabel="Days")
    if legend:
        g.add_legend()

plot_row("Rts", "Rt", legend=False)
plot_row("Is", "Infected", legend=True)
plot_row("Rs", "Recovered", legend=False)
```

```python
data = Homogeneous = R0Series[
    (R0Series["ProblemPlaceFraction"] == 0) &
    (R0Series["RunSets.Parameters.RunType"] == "difeq") &
    (R0Series["RunSets.Parameters.R0"] == 2.0) &
    (R0Series["RiskMean"] == 0.5) &
    (R0Series["RiskVariance"] == "low")
    
]
data = data.explode(["Ts", "Is", "Rs", "Rts"])
data["Susceptible"] = 1000.0 - data["Is"] - data["Rs"]
```

```python

g = sns.lineplot(data=data, y="Susceptible", x="Ts", label="Susceptible", color="black", linestyle="-")
g = sns.lineplot(data=data, y="Is", x="Ts", label="Infected", color="black", linestyle="--")
g = sns.lineplot(data=data, y="Rs", x="Ts", label="Recovered", color="black", linestyle=":")
g.set(ylabel="Individuals")
g.set(xlabel="Days")
```
