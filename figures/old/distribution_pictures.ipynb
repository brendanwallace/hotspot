```python
import numpy as np
import json
from scipy import stats
import matplotlib.pyplot as plt

DATA_LOCATION = '/Users/brendan/Documents/projects/risky_sir/go/data/extinctions/'

EXTINCTION_CUTOFF = 50

linestyles = ['-', '--', ':']
colors = ["blue", "magenta", "red"]
```

```python
# number of individuals
N = 1000
# chance of spreading per individual before recovering
a = 3/1000
```

```python
def compute_g(a, N=1000):
    upper_g = 1
    lower_g = 0
    for T in range(1000):
        g = (upper_g + lower_g)/2
        g_ = (1 - a + a * g)**N
        if g_ < g:
            upper_g = g
        else:
            lower_g = g
    return g

compute_g(8/1000)
```
